# Explore-exploit methods
## Multi-armed bandit
Cada bandido é uma máquina com uma alavanca.<br>
Toda vez que a alavanca é puxada **(pulled)**, ela retorna uma recompensa.<br>
Cada bandit retorna um valor de recompensa aleatório, mas que segue uma média $\mu$ própria. <br>
A **recompensa média populacional $\mu$** de cada bandit é desconhecida do jogador.<br><br>
O **objetivo do jogador** é puxar N alavancas e obter a maior recompensa possível, mesmo não conhecendo as recompensas.<br>

## Epsilon Greedy
Primeiro, se define um valor de **probabilidade $\epsilon$**.<br>
Para cada alavanca a ser puxada, deve-se escolher um bandit.<br>
A probabilidade $\epsilon$ define a proporção de escolhas aleatórias (com base em uma distribuição uniforme).<br>
A probabilidade **1 - $\epsilon$** define a proporção de escolhas do bandit com a maior recompensa estimada até então (para tentar maximizar o ganho).<br>
Parte-se do pressuposto que todas as recompensas são iguais a 0.<br>
A cada tentativa, atualiza-se a esperança de recompensa **(média amostral)** do bandit que foi puxado com base na recompensa dada.<br>

## Problema Proposto
Resolver para 3 bandits com **recompensas médias populacionais $\mu$** e **número de tentativas** dados pelo usuário.

```python
import numpy as np
import matplotlib.pyplot as plt
import Explore_Exploit_Methods
%matplotlib inline
```

```python
args = (1, 2, 3)
c100 = Explore_Exploit_Methods.experimento_epsgreedy(0.100, 200000, True, *args)
```

```python
N = 100000       # Número de tentativas em cada máquina
n = 30           # Número de máquinas para a média
args = (1, 2, 3)

c100 = [0]*N
c050 = [0]*N
c010 = [0]*N
c001 = [0]*N

for i in range(n):
    c100 += Explore_Exploit_Methods.experimento_epsgreedy(0.100, N, False, *args)/n
    c050 += Explore_Exploit_Methods.experimento_epsgreedy(0.050, N, False, *args)/n
    c010 += Explore_Exploit_Methods.experimento_epsgreedy(0.010, N, False, *args)/n
    c001 += Explore_Exploit_Methods.experimento_epsgreedy(0.001, N, False, *args)/n
    print('{:.1f}% '.format(100*(i+1)/n), end='')

fig = plt.figure(figsize=(7,4))
ax = fig.add_axes([0,0,1,1], xscale='log')
ax.plot(c100,label='ϵ = .100')
ax.plot(c050,label='ϵ = .050')
ax.plot(c010,label='ϵ = .010')
ax.plot(c001,label='ϵ = .001')
ax.legend(loc=4)
ax.set_title('Comparando valores de epsilon', fontSize=15)
```
