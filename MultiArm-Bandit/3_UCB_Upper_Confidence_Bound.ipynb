# Explore-exploit methods
## Multi-armed bandit
Cada bandido é uma máquina com uma alavanca.<br>
Toda vez que a alavanca é puxada **(pulled)**, ela retorna uma recompensa.<br>
Cada bandit retorna um valor de recompensa aleatório, mas que segue uma média $\mu$ própria. <br>
A **recompensa média populacional $\mu$** de cada bandit é desconhecida do jogador.<br><br>
O **objetivo do jogador** é puxar N alavancas e obter a maior recompensa possível, mesmo não conhecendo as recompensas.<br>

## UCB - Upper Confidence Bound
Método alternativo similar ao optmistic.<br>
Entretanto, a escolha do próximo bandit depende da sua média amostral e de quantas vezes tal bandit já foi testado.<br>

## Problema Proposto
Resolver para 3 bandits com **recompensas médias populacionais $\mu$**, **Limite Superior** e **número de tentativas** dados pelo usuário.

```python
import numpy as np
import matplotlib.pyplot as plt
import Explore_Exploit_Methods
%matplotlib inline
```

```python
args = (1, 2, 3)
cuc = Explore_Exploit_Methods.experimento_ucb(500000, True, *args)
```

```python
N = 150000       # Número de tentativas em cada máquina
n = 30           # Número de máquinas para a média
args = (1, 2, 3)

ceg = [0]*N
cuc = [0]*N
cop = [0]*N

for i in range(n):
    ceg += Explore_Exploit_Methods.experimento_epsgreedy(0.1, N, False, *args)/n
    cuc += Explore_Exploit_Methods.experimento_ucb(N, False, *args)/n
    cop += Explore_Exploit_Methods.experimento_optimistic(N, 10, False, *args)/n
    print('{:.1f}% '.format(100*(i+1)/n), end='')

fig = plt.figure(figsize=(7,4))
ax = fig.add_axes([0,0,1,1], xscale='log')
ax.plot(ceg,label='Epsilon Greedy')
ax.plot(cuc,label='UCB - Upper Confidence Bound')
ax.plot(cop,label='Optimstic Initial Values')
ax.legend(loc=4)
ax.set_title('Comparando métodos explore-exploit', fontSize=15)
```
